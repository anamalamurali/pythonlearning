```python
'''
1. Univariate Analysis -> 1 variable (Numerical {n} or Categorical {c})
2. Bi-Variate analysis -> 2 variables
NN --> both are numerical
CC --> both are categorical
NC/CN --> numeric and categorical

3. Multi variate analysis -> 3+ variables
NNN
CCC
NCN/NNC/CNN
CNC/CCN/NCC
 '''
```

```python
df.head()
```

```python
df['Genre'].unique()
```

```python
df['Genre'].nunique()
```

```python
cat_counts = df['Genre'].value_counts()
```

```python
cat_counts
```

```python
x_var = cat_counts.index
x_var
```

```python
y_var = cat_counts
```

```python
plt.bar(x_var,y_var)
```

```python
plt.figure(figsize=(12,8))
plt.bar(x_var,y_var)
```

```python
plt.bar(x_var,y_var)
plt.xticks(rotation=90, fontsize=12)
```

```python
plt.bar(x_var,y_var, width=0.2)
plt.xticks(rotation=90, fontsize=12)
plt.show()
```

```python
plt.bar(x_var,y_var)
plt.xticks(rotation=90, fontsize=12)
plt.xlabel("Genre", fontsize=10)
plt.ylabel("Count", fontsize=10)
plt.title("Count per Genre", fontsize=15)
plt.show()
```

```python
plt.bar(x_var,y_var, color='orange')
plt.xticks(rotation=90, fontsize=12)
plt.xlabel("Genre", fontsize=10)
plt.ylabel("Count", fontsize=10)
plt.title("Count per Genre", fontsize=15)
plt.show()
```
