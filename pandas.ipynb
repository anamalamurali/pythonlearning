# **Pandas:**

- data frame (df)
- df.head(n) -- no of top rows will return, by default will return 5 rows
- df.tail(n) -- no of bottom rows will return, by default will return 5 rows
- df.shape -- shape of array
- df["column_name"] -- returns all values in that column
- type(df) -- pandas.core.frame.DataFrame
- type(df["column_name"]) -- pandas.core.frame.Series
- df.info() - info about data frame
- String referred as object datatype in Pandas
- row-oriented apporach to create data frame
